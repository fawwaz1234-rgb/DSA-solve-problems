Recursion

```python
#printing the number 1 to N
n=5
def func(n):
    if n==0:
        return
    func(n-1)
    print(n)
func(n)
     
```

```python
#printing the N to 1
n=5
def func(n):
    if n==0:
        return
    print(n)
    func(n-1)
func(n)    
```

```python
#factorial of the number
n=5
def func(n):
    if n==0 or n==1:
        return 1
    else:
       return n*func(n-1)
func(n)    
```

```python
#sum of first N natural number
n=5
def func(n):
    if n==0:
        return 0
    else:
      return n+func(n-1)
func(n)  
```

```python
#print hello n times
n=5
def func(n):
    if n==0:
        return
    print("hello world")
    func(n-1)
func(n)    
```

```python
#counting numbers digit in number
n=12345

n = 12345

def func(n, count):
    if n == 0:
        return count

    return func(n//10, count + 1)

func(n, 0)

#whit no count
n=12345
def func(n):
    if n==0:
        return 0
    return 1+func(n//10)
print(func(n))
```

```python
#sum of the digit using recursion
n=527
def func(n):
    if n==0:
        return 0
    return (n%10)+func(n//10)
print(func(n))
```

```python
#find the product of the digit
n=234
def func(n):
    if n==0:
        return 1
    return (n%10)*func(n//10)
print(func(n))
```

```python
#printing even 1 to n numbeer
n=10
def func(n):
    if n==0:
        return n
    return func(n-2)
    
    
func(n)
```
