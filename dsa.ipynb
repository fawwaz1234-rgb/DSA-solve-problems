Recursion

```python
#printing the number 1 to N
n=5
def func(n):
    if n==0:
        return
    func(n-1)
    print(n)
func(n)
     
```

```python
#printing the N to 1
n=5
def func(n):
    if n==0:
        return
    print(n)
    func(n-1)
func(n)    
```

```python
#factorial of the number
n=5
def func(n):
    if n==0 or n==1:
        return 1
    else:
       return n*func(n-1)
func(n)    
```

```python
#sum of first N natural number
n=5
def func(n):
    if n==0:
        return 0
    else:
      return n+func(n-1)
func(n)  
```

```python
#print hello n times
n=5
def func(n):
    if n==0:
        return
    print("hello world")
    func(n-1)
func(n)    
```

level 2 recursion


```python
#count the number of digit in number
n=12345
def func(n):
    if n==0:
        return 0
    return 1+func(n//10)

func(n)

        
```

```python
#sum of the digit recrsilvely
n=12345
def func(n):
    if n==0:
        return 0
    return (n%10)+func(n//10)
func(n)
```

```python
#product of the number 
n=12345
def func(n):
    if n==0:
        return 1
    return (n%10)*func(n//10)
func(n)
```

```python
#even number 1 to n
n=120
def func(n):
    if n==0:
        return 
    func(n-1)
    if n%2==0:
        print(n)
func(n)    
```

```python
#odd number
n=120
def func(n):
    if n==0:
        return 
    func(n-1)
    if n%2!=0:
        print(n)
func(n)   
```

level 3 recusrion 

```python
#find the sum of the 1 to n number which are divisible by 3
n=10
def func(n):
    if n==0:
        return 0
    if n%3==0:
        return (n)+func(n-1)  
print(func(n))  
```
