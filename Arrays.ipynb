Array Manipulation

```python
#sawpping of the array
#reverse of the array
arr=[10,20,30,40,50]
left=0
right=(len(arr)-1)
while left<right:
    arr[left],arr[right]=arr[right],arr[left]
    left+=1
    right-=1
print(arr)    
```

```python
#print the array after every swap
arr=[10,10,5,41,24,7,6]
left=0
right=len(arr)-1#this one for comprassion
print(arr)
while left<right:
    arr[left],arr[right]=arr[right],arr[left]
    print(arr)
    left+=1
    right-=1
```

```python
#moving zero problems brute force metho
arr=[0,1,0,3,12]
arr2=[]
for i in arr:
    if i!=0:
        arr2.append(i)
while len(arr2)<len(arr):
        arr2.append(0)
print(arr2)
```

```python
#moving array O(1)
arr=[0,1,0,12,3]
w=0
for r in range(len(arr)):
    if arr[r]!=0:
        arr[w]=arr[r]
        w+=1
print(arr)        
    
```

```python
#removing duplicate of a sorted
arr=[1,1,2,2,3,4,4]
w=0
for r in range(len(arr)):
    if arr[r]!=arr[w]:
        w+=1
        arr[w]=arr[r]
        
print(arr)       
```
