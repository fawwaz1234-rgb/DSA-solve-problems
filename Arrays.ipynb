```python
#sum of the array element
arr=[15, 8, 22, 5, 18, 8]

sum=0
for num in arr:
    sum+=num
    
print(sum)
```

```python
#find the largest
largest=arr[0]

for i in arr:
    if i>largest:
        largest=i
print(largest)        
```

```python
#find the smallest
smallest=arr[0]

for i in arr:
    if i<smallest:
        smallest=i
print(smallest) 
```

```python
#count even numbers
count=0
for i in arr:
    if i%2==0:
        count+=1
print(count)      
```

```python
#count odd numbers
count=0
for i in arr:
    if i%2!=0:
        count+=1
print(count)
```

```python
#find the average of the array
sum=0
for i in arr:
    sum+=i
average=sum/len(arr)    
print(average)
```

```python
#find the index of 18
target=18
for i in range(len(arr)):
    if arr[i]==target:
        print(i,arr[i])
```

```python
#count how many times  8 occures
count=0
for i in range(len(arr)):
    if arr[i]==8:
        count+=1
print(count)
```

Searching in array 
time complexity o(n)

```python
#searcing 40 element index
arr2=[10,25,30,25,40,25,50]
target=40
for i in range(len(arr2)):
    if arr2[i]== target:
        print(i)
```

```python
#search for 100
target=100
found=False
for i in range(len(arr2)):
    if arr2[i]==target:
        print("Found")
        found= True
        break
if not found:
    print("not found")    
```

```python
#print all indices where 25 occurens
target=25
found=True
for i in range(len(arr2)):
    if arr2[i]==target:
        print(i)
        found= True
if not found :
    print("not found")        
```
